# Hệ số hòa hợp W theo Kendall

Hệ số hòa hợp W theo Kendall (Kendall's coefficient of concordance) là một trị số thống kê phi tham số do nhà thống kê Maurice Kendall (1907 - 1983) cho phép khảo sát về mức độ đồng thuận giữa k giám định viên trong việc đánh giá một đại lượng trên n trường hợp. Phương pháp thống kê này thường được sử dụng để khảo sát những kỹ thuật lâm sàng mà tính chính xác phụ thuộc vào yếu tố chủ quan do con người; thí dụ các điểm số là kết quả của việc đọc phim của các bác sĩ chẩn đoán hình ảnh.

Trong bài toán này, ta có $k$ giám định viên cùng đánh giá một thứ hạng $r$ cho $n$ đối tượng. Mỗi giám định viên $j$ đưa ra kết quả thứ hạng $r_{ij}$ cho mỗi đối tượng $i$. 

Tổng thứ hạng mà đối tượng $i$ nhận được là:

$$R_{i} = \sum_{k=1}^{k}r_{ij}$$

và thứ hạng trung bình trong mẫu khảo sát n đối tượng là:

$$\bar{R} = \frac{1}{n} \sum_{i=1}^{n}R_{i}$$

Tổng bình phương sai biệt S được ước tính như sau:

$$S = \sum_{i=1}^{n}\left ( R_{i} - \bar{R}\right )^2$$

Ngoài ra, thứ hạng có thể bằng nhau giữa nhiều giám định viên (ties), do đó đặt ra yêu cầu phải hiệu chinh cho hiện tượng này, thông qua một hệ số hiệu chỉnh $T_{j}$:

$$T_{j} = \sum_{i=1}^{g_{j}}\left ( t_{i}^{3} - t_{i}\right)$$

với $t_{i}$ là số thứ hạng bằng nhau trong nhóm thứ $i$ và $g_{j}$ là số nhóm thứ hạng bằng nhau trong tập hợp thứ hạng (từ 1 đến n) cho giám định viên $j$.

Công thức ước tính W có hiệu chỉnh thứ hạng bằng nhau:

$$W = \frac{12 \sum_{i=1}^{n}\left (R_{i}^2\right)- 3k^2n(n+1)^2}{k^2n\left (n^2-1 \right)- k(\sum_{j=1}^{k}\left ( T_{j} \right ))}$$

W có giá trị trong khoảng 0 (hoàn toàn không có sự hòa hợp) và 1 (hòa hợp tuyệt đối). 

# Kiểm định thống kê cho Kendall's W

Có hai phương pháp kiểm định thống kê được đề xuất cho bài toán khảo sát hệ số hòa hợp W, phương pháp thứ nhất được công bố vào năm 1990 bởi Jean Dickinson Gibbons, còn được gọi là kiểm định Q, với giả thuyết vô hiệu là không có sự hòa hợp (tương ứng với một hành vi xếp thứ hạng ngẫu nhiên không theo quy luật nào cả). Kiểm định này dựa vào trị số thống kê Q:

$$Q = k(n-1)W$$

với giả định là Q xấp xỉ phân phối $\chi^2$ với độ tự do $df = (n-1)$

Tuy nhiên, kiểm định Q này được đánh giá là quá bảo thủ. Do đó, vào năm 2014 tác giả Marco Marozzi đề xuất một phương pháp khác dựa vào phân phối F với độ tự do $\nu_{1} = n - 1 - (2/k)$ và $\nu_{2} = (k-1) \nu_{1}$

trị số thống kê F được ước tính bằng công thức:

$$F = \frac{W(k-1)}{1-W}$$

Marrozi cho rằng kiểm định F này hiệu quả tương đương với kiểm định hoán vị và nên dùng khi giá trị k thấp.

# Liên hệ giữa Kendall's W và kiểm định Friedman

Thực ra, có liên hệ chặt chẽ giữa hệ số hòa hợp W theo Kendall và kiểm định phi tham số Friedman.

Như ta đều biết: Friedman test do nhà thống kê Milton Friedman (1912 –2006) công bố vào năm 1940, là một kiểm định phi tham số thay thế cho ANOVA phép đo lặp lại. 

Do trong thực hành, hai phương pháp Kendall's W và Friedman test được dùng trong những bối cảnh hoàn toàn khác biệt (khảo sát mức độ tương hợp giữa nhiều bác sĩ, và so sánh trung bình một đại lượng được đo ở nhiều thời điểm/điều kiện khác nhau), nên nhiều bạn vẫn nghĩ rằng chúng là 2 kỹ thuật khác nhau.

Tuy 2 bài toán này có vẻ khác biệt, nhưng chúng có cùng bản chất: cả 2 đều có 1 biến phụ thuộc là đại lượng cần khảo sát (thí dụ thang điểm chẩn đoán hình ảnh, t lại chú ý rằng: chính vì nó không liên tục, nên ta đã không dùng được ANOVA mà phải dùng Friedman test), 1 yếu tố điều kiện (k giám định viên), và mục tiêu đều là chứng minh có sự khác biệt (hoặc ngược lại: tương đồng) giữa các giám định viên.

Khi xem lại quy trình, ta sẽ thấy Kendall's W chính là bước trung gian để thực hiện được kiểm định F trong Friedman test,cả 2 kỹ thuật đều bắt đầu bằng việc xếp hạng, sau đó ta có công thức tính F xấp xỉ trong Friedman test:

$$F = \frac{12}{nk(k+1)}\sum_{j=1}^{k}R_{j}^{2} - 3n(k+1)$$

với k là số điều kiện, n là số trường hợp, $R_{j}^{2}$ là bình phương tổng thứ hạng cho một phân nhóm j (j=1,2,3...k)

Ta nhận ra rằng giá trị F theo phương pháp xấp xỉ trong Friedman test chính là Q trong Q test của Kendall's W; và cả 2 đều dựa vào phân phối $\chi^2$ với độ tự do k-1.

1 object thuộc class này có thể được khởi tạo bằng 3 cách:

1) Sử dụng bảng dọc (long format), lúc này dùng những arguments như sau:

df_format = 'long', var_name = None, subj = tên cột chỉ mã số định danh cá thể, cond = tên cột chỉ yếu tố can thiệp (lượt đo, thời gian, điều kiện đo, giám định viên), target = tên cột chỉ biến kết quả Y

2) Sử dụng bảng ngang (wide format), với arguments như sau:

df_format = 'wide', var_name = 1 cái tên tùy thích chỉ yếu tố can thiệp (lượt đo, thời gian, điều kiện đo, giám định viên/bác sĩ), subj = None, cond = 1 list chứa tên k cột phân nhóm cần khảo sát (lượt đo, điều kiện, phân nhóm can thiệp, mã số k giám định viên/bác sĩ), target = tên biến kết quả Y, tùy ý.

3) Sử dụng công thức trên 1 bảng dọc, thông qua class method 'from_formula': Y ~ C | S với Y là tên biến kết quả, C là tên yếu tố can thiệp (phân nhóm), S là tên của yếu tố cá thể (thí dụ Patient, Subject) trong dataframe long format.

Hàm init sẽ tạo ra 1 object với 3 thuộc tính là 1 dataframe long format, 1 dataframe wide format (vì wide format sẽ tiện dụng để làm F test, trong khi long format tiện dụng để làm post hoc test), và 1 list chứa tên cond, subj,target.

Class này có 5 method, 

method testing kết hợp cả 2 quy trình (1) tính Kendall's W kèm theo Q test theo Gibbon (tương đương kiểm định Chi2 cho F xấp xỉ) và Marrozi F test; (2) Kiểm định F của Iman-Davenport cho F exact của Friedman test. 

và 4 loại post-hoc test khác nhau.

```python
import warnings
warnings.filterwarnings('ignore')

# Công cụ
import pandas as pd
import numpy as np

# Đồ họa
import seaborn as sns
import matplotlib.pyplot as plt

# Thống kê
from scipy.stats import f as f_test, shapiro as sw

from itertools import combinations

import scipy.stats as ss

from statsmodels.sandbox.stats.multicomp import multipletests
from statsmodels.stats.libqsturng import psturng
```

```python
class Friedman_test:

    """Class thực hiện kiểm định Friedman, Kendall's W và
    các kiểm định post-hoc tương ứng"""

    def __init__(self, df: pd.DataFrame, var_name: str=None, subj:str=None,cond = None,target:str=None, df_format: str='long', ):
        
        df = df.copy()
        if df_format == 'wide':
            df[subj] = df.index
            ldf = pd.melt(df, id_vars = [subj], 
                          value_vars = cond,
                          value_name = target,
                          var_name = var_name
                         )
            ldf[subj]= ldf[subj].astype('object')
            
            wdf = df.drop([subj], axis = 1)
            cond = var_name
                        
        if df_format == 'long':
            wdf = df.pivot(index = subj, 
                          columns = cond,
                          values = target)
            
            ldf = df[[subj, cond, target]]

        self.ldf = ldf.dropna(how = 'any', axis = 0)
        self.wdf = wdf.dropna(how = 'any', axis = 0)
        
        self.params = [cond, subj, target]

    @classmethod
    def from_formula(cls, df: pd.DataFrame, formula = 'Y ~ C | G'):
        target = formula.split('~')[0].strip()
        cond = formula.split('~')[1].split('|')[0].strip()
        subj = formula.split('~')[1].split('|')[1].strip()
        
        return cls(df = df, 
                   df_format = 'long', 
                   var_name = cond,
                   subj=subj, 
                   cond=cond, 
                   target=target)

    def __repr__(self):
        return f"Kiểm định Friedman với k={self.wdf.shape[1]}, n={self.wdf.shape[0]}"

    def testing(self, alpha: float = 0.05):
        data = self.wdf

        stat = pd.DataFrame(index = ["Kendall' W", 
                                     "Gibbon's Q test",
                                     "Chi2 cho F exact",
                                     "Iman-Davenport F-test",
                                     "Marrozi F-test"])

        X = data.values
        conds = list(data.columns)
        n,k = data.shape

        rank_mat = np.zeros(X.shape)

        for i in range(n):
            rank_mat[i] = ss.rankdata(X[i, :], method = 'average')

        ssb = (rank_mat.sum(axis=0)**2).sum()
        
        # Hiệu chỉnh thứ hạng bằng nhau
        ties = 0
        for i in range(n):
            replist, repnum = ss.find_repeats(X[i])
            for t in repnum:
                ties += (t**3 - t)

        # Kendall's W
        W = (12 * ssb - 3 * n * n * k * (k + 1) * (k + 1)) / (n * n * k * (k - 1) * (k + 1) - n * ties)
        
        # Chisquared test

        # Q statistic
        Q = n * (k - 1) * W
                
        # Độ tự do 
        dof = k - 1
        p1 = ss.chi2.sf(Q, dof)

        # F chính xác
        rj = rank_mat.mean(axis = 0)
        rm = rank_mat.mean()
        SST = n*((rj - rm)**2).sum()
        SSE = ((rank_mat-rm)**2).sum()/(n*(k-1))
        
        F_exact = SST/SSE

        # Hiệu chỉnh ties
        c = 1 - ties / float(k * (k * k - 1) * n)
        F_exact /= c

        p2 = ss.chi2.sf(F_exact, dof)

        # Iman-Davenport F test
        Fc = ((n - 1)*Q)/(n*(k-1) - Q)
        dof_1 = k-1
        dof_2 = (k-1)*(n-1)
        p3 = f_test.sf(Fc, dof_1, dof_2)

        # Marozzi F test
        Fm = W * (n - 1) / (1 - W)

        mdof_1 = k - 1 - 2 / n
        mdof_2 = (n - 1) * mdof_1
        p4 = f_test.sf(Fm, mdof_1, mdof_2)

        stat['Trị số'] = ['W','Q','F-exact','Fc','F']
        stat['Ước tính'] = [W,Q,F_exact,Fc,Fm]
        stat['Độ tự do'] = [str(''), str(dof), str(dof), f"({dof_1},{dof_2})", f"({mdof_1:.2f},{mdof_2:.2f})"]
        stat['Giá trị p'] = [None,p1,p2,p3,p4]
        stat['Phủ định H0'] = ['Có thể' if i else 'Không thể' for i in stat['Giá trị p'] < alpha]
        
        stat['Phủ định H0'][0] = ''
        stat['Giá trị p'][0] = None
        stat.index.name = 'Friedman'

        return stat
    
    def Nemenyi_test(self):
        df = self.ldf
        cond, subj, target = self.params
        
        groups = df[cond].unique()
        k = groups.size
        n = df[subj].unique().size
        
        df['Rank'] = df.groupby(subj)[target].rank()
        
        R = df.groupby(cond)['Rank'].mean()
        
        vs = np.zeros((k, k))
        combs = combinations(range(k), 2)
        
        def nemenyi_stats(i, j):
            dif = np.abs(R[groups[i]] - R[groups[j]])
            qval = dif / np.sqrt(k * (k + 1.) / (6. * n))
            return qval
        
        tri_upper = np.triu_indices(vs.shape[0], 1)
        tri_lower = np.tril_indices(vs.shape[0],-1)
        vs[:, :] = 0

        for i, j in combs:
            vs[i, j] = nemenyi_stats(i, j)
            
        vs *= np.sqrt(2.)
        vs[tri_upper] = psturng(vs[tri_upper], k, np.inf)
        vs[tri_lower] = vs.T[tri_lower]
        np.fill_diagonal(vs, 1)

        return pd.DataFrame(vs, index=groups, columns=groups)
    
    def Conover_test(self, p_adjust = 'bonferroni'):
        
        df = self.ldf
        cond, subj, target = self.params
        
        groups = df[cond].unique()
        k = groups.size
        n = df[subj].unique().size
        
        df['Rank'] = df.groupby(subj)[target].rank()
        
        R = df.groupby(cond)['Rank'].sum()
        
        def posthoc_conover(i, j):
            dif = np.abs(R.loc[groups[i]] - R.loc[groups[j]])
            tval = dif / np.sqrt(A) / np.sqrt(B)
            pval = 2. * ss.t.sf(np.abs(tval), df=(m*n*k - k - n + 1))
            return pval
        
        A1 = (df['Rank'] ** 2).sum()
        m = 1
        S2 = m/(m*k - 1.) * (A1 - m*k*n*(m*k + 1.)**2./4.)
        T2 = 1. / S2 * (np.sum(R) - n * m * ((m * k + 1.) / 2.)**2.)
        A = S2 * (2. * n * (m * k - 1.)) / (m * n * k - k - n + 1.)
        B = 1. - T2 / (n * (m * k - 1.))
        
        vs = np.zeros((k, k))
        combs = combinations(range(k), 2)

        tri_upper = np.triu_indices(vs.shape[0], 1)
        tri_lower = np.tril_indices(vs.shape[0], -1)
        vs[:, :] = 0
        
        for i, j in combs:
            vs[i, j] = posthoc_conover(i, j)

        if p_adjust:
            vs[tri_upper] = multipletests(vs[tri_upper], method=p_adjust)[1]

        vs[tri_lower] = vs.T[tri_lower]
        np.fill_diagonal(vs, 1)
        
        return pd.DataFrame(vs, index=groups, columns=groups)
    
    def Siegel_Castellan(self, p_adjust = 'bonferroni'):
        
        df = self.ldf
        cond, subj, target = self.params
        
        groups = df[cond].unique()
        k = groups.size
        n = df[subj].unique().size
        
        df['Rank'] = df.groupby(subj)[target].rank()
        
        R = df.groupby(cond)['Rank'].mean()
        
        def posthoc_SC(i, j):
            dif = np.abs(R[groups[i]] - R[groups[j]])
            zval = dif / np.sqrt(k * (k + 1.) / (6. * n))
            return zval
        
        vs = np.zeros((k, k), dtype=np.float)
        combs = combinations(range(k), 2)

        tri_upper = np.triu_indices(vs.shape[0], 1)
        tri_lower = np.tril_indices(vs.shape[0],-1)
        vs[:, :] = 0

        for i, j in combs:
            vs[i, j] = posthoc_SC(i, j)
            
        vs = 2. * ss.norm.sf(np.abs(vs))
        vs[vs > 1] = 1
        
        if p_adjust:
            vs[tri_upper] = multipletests(vs[tri_upper], method=p_adjust)[1]

        vs[tri_lower] = vs.T[tri_lower]
        np.fill_diagonal(vs, 1)

        return pd.DataFrame(vs, index=groups, columns=groups)
    
    def Miller_Friedman(self):
        
        df = self.ldf
        cond, subj, target = self.params
        
        groups = df[cond].unique()
        k = groups.size
        n = df[subj].unique().size
        
        df['Rank'] = df.groupby(subj)[target].rank()
        
        R = df.groupby(cond)['Rank'].mean()
        
        def posthoc_Miller(i, j):
            dif = np.abs(R[groups[i]] - R[groups[j]])
            qval = dif / np.sqrt(k * (k + 1.) / (6. * n))
            return qval
        
        vs = np.zeros((k, k), dtype=np.float)
        combs = combinations(range(k), 2)

        tri_upper = np.triu_indices(vs.shape[0], 1)
        tri_lower = np.tril_indices(vs.shape[0], -1)
        vs[:, :] = 0

        for i, j in combs:
            vs[i, j] = posthoc_Miller(i, j)

        vs = vs ** 2
        vs = ss.chi2.sf(vs, k - 1)
        
        vs[tri_lower] = vs.T[tri_lower]
        np.fill_diagonal(vs, 1)

        return pd.DataFrame(vs, index=groups, columns=groups)
```
